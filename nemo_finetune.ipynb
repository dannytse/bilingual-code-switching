```python
import torch
import pandas as pd
import datasets
import json
from nemo.collections.asr.models import EncDecRNNTModel
import os
import re
import soundfile as sf
from langdetect import detect, DetectorFactory
from langdetect.lang_detect_exception import LangDetectException
from tqdm import tqdm

# Load the ASCEND dataset
train_set = datasets.load_dataset("CAiRE/ASCEND", split="train")
val_set = datasets.load_dataset("CAiRE/ASCEND", split="validation")
test_set = datasets.load_dataset("CAiRE/ASCEND", split="test")

DetectorFactory.seed = 0

def split_text_by_language(text):
    en_pattern = re.compile(r'[A-Za-z]+')
    zh_pattern = re.compile(r'[\u4e00-\u9fff]+')

    chunks = []
    current_chunk = ""
    current_language = "en"

    for char in text:
        if re.match(en_pattern, char):
            if current_language != "en":
                if current_chunk:
                    chunks.append((current_chunk, current_language))
                current_chunk = char
                current_language = "en"
            else:
                current_chunk += char
        elif re.match(zh_pattern, char):
            if current_language != "zh":
                if current_chunk:
                    chunks.append((current_chunk, current_language))
                current_chunk = char
                current_language = "zh"
            else:
                current_chunk += char
        else:
            current_chunk += char

    if current_chunk:
        chunks.append((current_chunk, current_language))

    return chunks

os.makedirs("train_audio_files", exist_ok=True)
os.makedirs("val_audio_files", exist_ok=True)
os.makedirs("test_audio_files", exist_ok=True)

# Save audio files and convert dataset to NeMo's manifest format
def save_audio_and_convert_to_manifest(dataset, output_file, audio_dir):
    manifest_data = []
    with tqdm(total=len(dataset), position=0, leave=True) as pbar:
        for example in dataset:
            audio_array = example["audio"]["array"]
            sampling_rate = example["audio"]["sampling_rate"]
            audio_file_name = example["audio"]["path"]
            audio_file_path = os.path.join(audio_dir, audio_file_name)

            sf.write(audio_file_path, audio_array, sampling_rate)

            text = example["transcription"]
            duration = example["duration"]
            segments = split_text_by_language(text)
            texts = []
            for segment, lang in segments:
                texts.append({"str": segment, "lang": lang})

            if duration is not None:
                manifest_data.append({
                    "audio_filepath": audio_file_path,
                    "duration": duration,
                    "text": texts
                })
            pbar.update()

    with open(output_file, 'w') as f:
        for item in manifest_data:
            f.write(f"{json.dumps(item)}\n")

train_manifest = os.path.abspath("train_manifest.json")
val_manifest = os.path.abspath("val_manifest.json")
test_manifest = os.path.abspath("test_manifest.json")

save_audio_and_convert_to_manifest(train_set, train_manifest, "train_audio_files")
save_audio_and_convert_to_manifest(val_set, val_manifest, "val_audio_files")
save_audio_and_convert_to_manifest(test_set, test_manifest, "test_audio_files")
```

```python
import os
from nemo.collections.asr.models import EncDecRNNTBPEModel
from pytorch_lightning import Trainer
import torch

model_path = os.path.abspath("./stt_enzh_fastconformer_transducer_large_codesw/stt_enzh_fastconformer_transducer_large_codesw.nemo")
model = EncDecRNNTBPEModel.restore_from(model_path)

train_manifest = "train_manifest.json"
val_manifest = "val_manifest.json"
test_manifest = "test_manifest.json"

model.cfg.train_ds.manifest_filepath = [train_manifest]
model.cfg.train_ds.batch_size = 16
model.cfg.train_ds.is_tarred = False  # Ensure it is not using tarred datasets
model.cfg.train_ds.tarred_audio_filepaths = None
model.cfg.train_ds.is_concat = False  # Ensure concatenation is not used
model.cfg.train_ds.concat_sampling_technique = None
model.cfg.train_ds.concat_sampling_probabilities = None
model.cfg.validation_ds.manifest_filepath = [val_manifest]
model.cfg.validation_ds.batch_size = 8
# model.cfg.validation_ds.is_tarred = False
# model.cfg.validation_ds.tarred_audio_filepaths = None
model.cfg.test_ds.manifest_filepath = [test_manifest]
model.cfg.test_ds.batch_size = 8
# model.cfg.test_ds.is_tarred = False
# model.cfg.test_ds.tarred_audio_filepaths = None

model.cfg.optim.lr = 0.0001

print(f"Setting up training data from {model.cfg.train_ds.manifest_filepath}")
model.setup_training_data(train_data_config=model.cfg.train_ds)

print(f"Setting up validation data from {model.cfg.validation_ds.manifest_filepath}")
model.setup_validation_data(val_data_config=model.cfg.validation_ds)

# Define trainer
trainer = Trainer(
    max_epochs=10,
    precision=32,
)

print("Starting training...")
trainer.fit(model, model.train_dataloader(), model.val_dataloader())

print("Training completed.")

model.save_to("finetuned_stt_enzh_fastconformer_transducer_large_codesw.nemo")
```

## Evaluation

```python
import chinese_converter
import evaluate
import re

metric = evaluate.load("wer")

def insert_space_in_code_switched_text(text):
    text = text.lower()
    # Regular expression to match Chinese characters.
    chinese_char_pattern = r'[\u4e00-\u9fff]'

    # Insert space before and after each Chinese character.
    spaced_text = re.sub(f'({chinese_char_pattern})', r' \1 ', text)

    # Remove any extra spaces added by the previous step.
    normalized_text = re.sub(r'\s+', ' ', spaced_text)
    normalized_text = normalized_text.strip().replace("  ", " ")
    return normalized_text


def calculate_MER(predictions, references):
    predictions = [insert_space_in_code_switched_text(chinese_converter.to_simplified(string[0])) for string in predictions]
    refs_concatenated = []
    for ref in references:
        ref_concatenated = "".join([x["str"] for x in ref])
        refs_concatenated.append(ref_concatenated)
    references = [insert_space_in_code_switched_text(chinese_converter.to_simplified(string)) for string in refs_concatenated]
    return 100 *metric.compute(predictions=predictions, references=references)
```

```python
def generate_predictions_from_manifest(model, manifest_path):
    with open(manifest_path, 'r', encoding='utf-8') as f:
        eval_data = [json.loads(line.strip()) for line in f]
    predictions = []
    references = []
    for data in eval_data:
        transcript = model.transcribe([data["audio_filepath"]])[0]
        predictions.append(transcript)
        references.append(data["text"])
    return predictions, references

# Evaluate pre-finetuning
print("Evaluating pre-finetuning...")
model = EncDecRNNTBPEModel.restore_from("/content/stt_enzh_fastconformer_transducer_large_codesw/stt_enzh_fastconformer_transducer_large_codesw.nemo")
predictions_pre, references_pre = generate_predictions_from_manifest(model, val_manifest)
mer_pre = calculate_MER(predictions_pre, references_pre)

model_finetuned = EncDecRNNTBPEModel.restore_from("finetuned_stt_enzh_fastconformer_transducer_large_codesw.nemo")
# Evaluate post-finetuning
print("Evaluating post-finetuning...")
predictions_post, references_post = generate_predictions_from_manifest(model_finetuned, val_manifest)
mer_post = calculate_MER(predictions_post, references_post)
```
